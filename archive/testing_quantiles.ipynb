```r
require(quantmod)
require(Rcpp)
require(microbenchmark)
```

```r
sourceCpp('quantile.cpp')
```

```r
run_quantileR = function(x, n, probs) {
    out = c( rep(NA,(n-1)), sapply(n:length(x), function(i) quantile(x[(i- (n-1) ):i], probs) ))
    as.vector(out)
}


# basic test

n = 1000
probs = 0.9
x = runif(100000)

print(summary(microbenchmark(
	run_quantile(x, n, probs),
	times = 2
)),0)
```
